```python
import pyautogui
import time
from datetime import datetime
from pathlib import Path
import win32com.client as win32
import shutil


pyautogui.PAUSE = 1
def mes():
    mes = str(int(datetime.today().strftime('%m'))-1)
    if mes == '1':
        mes,img_mes ='janeiro','janeiro.png'
    if mes == '2':
        mes,img_mes ='fevereiro','fevereiro.png'
    if mes == '3':
        mes,img_mes ='marco','marco.png'
    if mes == '4':
        mes,img_mes ='abril','abril.png'
    if mes == '5':
        mes,img_mes ='maio','maio.png'
    if mes == '6':
        mes,img_mes ='junho','junho.png'
    if mes == '7':
        mes,img_mes ='julho','julho.png'
    if mes == '8':
        mes,img_mes ='agosto','agosto.png'
    if mes == '9':
        mes,img_mes ='setembro','setembro.png'
    if mes == '10':
        mes,img_mes ='outubro','outubro.png'
    if mes == '11':
        mes,img_mes ='novembro','novembro.png'
    if mes == '12':
        mes,img_mes ='dezembo','dezembro.png'
        
    return (mes,str(img_mes))

def acessar_url(url):
    pyautogui.press('win')
    pyautogui.write('edge')
    pyautogui.press('enter')
    pyautogui.write(url)
    pyautogui.press('enter')
    
def select_cnpj(cnpj):
    time.sleep(2)
    pyautogui.write(cnpj)
    pyautogui.press('enter')
    
    
    

def selecionar(**kwargs):
 
    for imagem in kwargs.values():
        #esperar carregar o google
        while not pyautogui.locateOnScreen(imagem):
            time.sleep(1)

        # localizar a imagem -> vai te dar 4 informações: posicao x, posicao y, largura e altura
        x, y, largura, altura = pyautogui.locateOnScreen(imagem)
        # clicar no meio da imagem
        pyautogui.click(x + largura/2, y + altura/2)
        
def selecionar_mes_emissao(img_mes):
        # localizar a imagem -> vai te dar 4 informações: posicao x, posicao y, largura e altura
        x, y, largura, altura = pyautogui.locateOnScreen(img_mes)
        # clicar no meio da imagem
        pyautogui.click(x + 2, y + altura/2)
        
def criar_pasta(nome,mes):
    #caminho aonde ficam todos os clienntes
    caminho=Path(rf'C:\Users\alexs\PycharmProjects\AUTOMAÇÃO MEI\DAS')

    #nome da pasta do cliente
    pasta_cliente = Path(fr'{caminho}\{nome}')

    #itens é os arquivos no caminho
    itens = caminho.iterdir()


    #se o nome do caminho é igual ao caminho da pasta do cliente
    if  str(pasta_cliente) in itens:
        pass
    else:
        try:
            Path(fr'C:\Users\alexs\PycharmProjects\AUTOMAÇÃO MEI\DAS\{nome}').mkdir()

        except(FileExistsError):
            print(f'{pasta_cliente} já exite no local')


    pasta_mes = fr'{caminho}\{nome}\{mes}'
    novo_caminho = Path(fr'{caminho}\{nome}')
    itens = novo_caminho.iterdir()

    if  str(pasta_mes) in itens:
        pass
    else:
        try:
            Path(fr'C:\Users\alexs\PycharmProjects\AUTOMAÇÃO MEI\DAS\{nome}\{mes}').mkdir()

        except(FileExistsError):
            print(f'{pasta_mes} já exite no local')
    
        
def mover_para_pasta_das(nome,mes):
    caminho = Path(r"C:\Users\alexs\Downloads")
    arquivos = Path.iterdir(caminho)
    novo_caminho = Path(fr'C:\Users\alexs\PycharmProjects\AUTOMAÇÃO MEI\DAS\{nome}\{mes}')
    
    for arquivo in arquivos:
        if '.pdf' and 'DAS' in arquivo.name :
            das = arquivo.name
            try:
                Path(fr'{caminho}\{das}').rename(fr'{novo_caminho}\DAS-{nome}-COMPETENCIA {mes}.pdf')
            except(FileNotFoundError):
                criar_pasta(nome,mes)
                time.sleep(3)
                Path(fr'{caminho}\{das}').rename(fr'{novo_caminho}\DAS-{nome}-COMPETENCIA {mes}.pdf')
                
            
def enviar_email(nome,mes,endereco):
    # antes de usar limpar os rascunhos salvos

    # criar a integração com o outlook
    outlook = win32.Dispatch('outlook.application')

    # criar um email
    email = outlook.CreateItem(0)

    # configurar as informações do seu e-mail
    email.To = endereco
    print(email.To)
    email.Subject = rf'{nome} - DAS MEI {mes}/2023'
    email.HTMLBody = rf'Bom dia! Segue em anexo seu boleto mensal para pagamento do imposto do MEI.'
    attachment  = fr'C:\Users\alexs\PycharmProjects\AUTOMAÇÃO MEI\DAS\{nome}\{mes}\DAS-{nome}-COMPETENCIA {mes}.pdf'
    email.Attachments.Add(attachment)
    email.Send()
    print(f"Email Enviado para {nome}")
    
    
class cliente:
    def __init__(self,nome,cnpj,email):
        self.nome=nome
        self.cnpj=cnpj
        self.email=email
    """defina:
nome;
cnpj;
e-mail;"""

   
```

```python


pyautogui.PAUSE = 1

url = 'https://www8.receita.fazenda.gov.br/SimplesNacional/Aplicacoes/ATSPO/pgmei.app/Identificacao/'
acessar_url(url)
clientes = []
clientes.append(cliente('NOME','cnpj','E-MAIL'))


for cliente in clientes:
    
    select_cnpj(cliente.cnpj)

    selecionar(entrar_emitir_guia = 'Emitir Guia de Pagamento (DAS).png',informar_ano = 'informe ano calendario.png', ano = '2023.png', ok = 'ok.png')

    try:
        mes,img_mes = mes()

    except:
        img_mes = str(img_mes)

    selecionar_mes_emissao(img_mes)
    time.sleep(2)
    pyautogui.hotkey('fn','pgdn')

    selecionar(apurar_gerar = 'apurar_gerar.png',imprimir = 'imprimir.png')
    time.sleep(5)
    mover_para_pasta_das(cliente.nome,mes)
    time.sleep(5)
    enviar_email(cliente.nome,mes,cliente.email)
    time.sleep(5)
```
